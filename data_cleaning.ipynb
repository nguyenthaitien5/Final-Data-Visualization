```python
import re
```

```python
def strip_str(s):
#     postfix = s.strip(' ')[-1]
#     if postfix != 'm':
#         s = np.NaN
#         return s
    if s.count('m') != 1:
        return np.NaN
    x = re.search("[+-]?([0-9]*[,])?[0-9]+", s)
    if(x != None):
        s = float(s[x.start(): x.end()].replace(',', '.'))
    else:
        s = np.NaN
    return s
    
df['width'] = df['width'].apply(strip_str)
df['height'] = df['height'].apply(strip_str)
df['area'] = df['area'].apply(strip_str)
df['street_in_front_of_house'] = df['street_in_front_of_house'].apply(strip_str)
df.head(10)
```

```python
print(strip_str('3,6 tỷ'))
```

```python
def find_str(s):
    x = re.search("[+-]?([0-9]*[,])?[0-9]+", s)
    if(x != None):
        # x.end() là index của giá trị khoảng trắng sau số cuối cùng, nên cần cộng 1 để tính từ ký tự tiếp theo
        s = s[x.end() + 1 : len(s)]
    return s
def strip_str_price(s):
    x = re.search("[+-]?([0-9]*[,])?[0-9]+", s)
    if(x != None):
        s = float(s[x.start(): x.end()].replace(',', '.'))
    else:
        s = np.NaN
    return s
df['postfix_price'] = df['price'].apply(find_str)
df['price'] = df['price'].apply(strip_str_price)
df.head(100)
```

```python
print(f"Unique FN: {df['postfix_price'].unique()}")
```

```python
def replace_price(row):
    if row['postfix_price'] == "tỷ ":
        return row['price']
    elif  row['postfix_price'] == "triệu ":
        return row['price'] /1000
    else :
        return np.NaN
    
df['price'] = df.apply (lambda row: replace_price(row), axis=1)
# df['price'] = df['price'].apply(replace_price)
```

```python
df.head(10)
```

```python
df.dropna(inplace=True)
df.shape
```

```python
def split_address(row):
    x = re.split(',', row['address'])
    district = x[-2].strip()
    province = x[-1].strip()
    row['district'] = district
    row['province'] = province
    return row


df = df.apply(lambda row: split_address(row), axis=1)
```

```python
df.head(10)
```

```python
def replace_date(s):
    x = re.search("(0[1-9]|[12][0-9]|3[01])[- /.](0[1-9]|1[012])[- /.](19|20)\d\d",s)
    if (x != None):
        s = s[x.start() : x.end()]
    return s
    
df['start_date'] = pd.to_datetime(df['start_date'].apply(replace_date) ,format= "%d/%m/%Y")
```
