```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Raw_DF.csv', encoding='utf-8')
df.head(10)
```

```python
df.shape
```

```python
print(f"Unique FN: {df['direction'].unique()}")
```

```python
df.drop(columns=['description', 'address', 'end_date', 'direction', 'law'], inplace=True)
```

```python
df.head(10)
```

```python
# df.replace('_', np.NaN, inplace=True)
df.replace('---', np.NaN, inplace=True)
df.head(10)
```

```python
df.dropna(inplace=True)
df.shape
```

```python
df.info()
```

```python
import re
```

```python
def strip_str(s):
    x = re.search("[+-]?([0-9]*[,])?[0-9]+", s)
    if(x != None):
        s = s[x.start(): x.end()]
    return s
    
df['width'] = df['width'].apply(strip_str)
df['height'] = df['height'].apply(strip_str)
df['area'] = df['area'].apply(strip_str)
df['street_in_front_of_house'] = df['street_in_front_of_house'].apply(strip_str)
df.head(10)
```

```python
print(strip_str('3,6 tỷ'))
```

```python
def find_str(s):
    x = re.search("[+-]?([0-9]*[,])?[0-9]+", s)
    if(x != None):
        # x.end() là index của giá trị khoảng trắng sau số cuối cùng, nên cần cộng 1 để tính từ ký tự tiếp theo
        s = s[x.end() + 1 : len(s)]
    return s

df['postfix_price'] = df['price'].apply(find_str)
df['price'] = df['price'].apply(strip_str)
df.head(100)
```

```python
print(f"Unique FN: {df['postfix_price'].unique()}")
```

```python
def replace_price(row):
    if row['postfix_price'] == "tỷ ":
        return strip_str(row['price']).replace(",",".")
    elif  row['postfix_price'] == "triệu ":
        return float(strip_str(row['price']).replace(",","."))/1000
    else :
        return np.NaN
    
df['price'] = df.apply (lambda row: replace_price(row), axis=1)
```
